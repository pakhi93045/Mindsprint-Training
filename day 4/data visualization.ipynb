#matplotlib

```python
import matplotlib.pyplot as plt
print(plt.__package__)
import numpy as np
import pandas as pd
```

```python
plt.figure(figsize=(10,6))
```

```python
plt.plot([1,2,3], [4,5,6])
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,25,30,40]

plt.plot(x,y,marker='o', linestyle='--', color='g')

plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.title("Simple Line PLot")
plt.grid(True)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,25,30,40]

plt.scatter(x,y,marker='*', color='grey')

plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.title("Scatter PLot")

plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.bar(categories, values, color='c')
plt.xlabel("Categories")
plt.ylabel("Label")
plt.title("Bar Chart Example")

plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='pink', edgecolor='white')
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
sizes = [20, 30, 25, 25]
l = ["Python", "Java", "C++", "JavaScript"]
colors = ['brown', 'pink', 'red', 'violet']
plt.pie(sizes, labels=l, colors=colors, autopct='%1.1f%%')
plt.show()
```

```python
sizes = [20, 30, 25, 25]
plt.pie(sizes)
plt.show()
```

#activity

```python
sales = np.array([1200, 4567, 2435, 7834, 3798, 8355, 6209, 5467, 3252, 8920])
days=np.arange(1,11) #it will create array of 1 to 10
print(sales)
```

```python
print("Total sales: ",np.sum(sales))
average_sales = np.average(sales)
print("Average sales: ",average_sales)
print("Highest sales: ",np.max(sales))
print("Lowest sales: ",np.min(sales))
```

```python
plt.figure(figsize=(8,5))
plt.plot(days, sales, marker='*', linestyle='--', color='brown', label="Sales")

plt.axhline(average_sales, color='grey', linestyle='-', label="Average Sales")

plt.xlabel("Days")
plt.ylabel("Sales Revenue ($)")
plt.title("Sales Trend Over 10 days")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(days, sales, color='c', label='Sales')

plt.xlabel("Days")
plt.ylabel("Sales Revenue ($)")
plt.title("Sales per Day")
plt.legend(loc="upper right")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

          
```

```python
plt.figure(figsize=(6,6))
plt.pie(sales, labels=days, colors=plt.cm.Paired.colors, autopct='%1.1f%%')   #plt.cm.Paired.colors - it is package for colors, if we dont want to give colors it will assign colors randomly
plt.title("sales contribution per day")
plt.show()
```
